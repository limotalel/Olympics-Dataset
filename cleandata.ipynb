```python
import regex as re
date_pattern = r'\b(\d{1,2}\s+[A-Za-z]{3,9}\s+\d{4}|[A-Za-z]{3,9}\s+\d{1,2},?\s+\d{4}|\d{4}-\d{2}-\d{2}|\d{1,2}/\d{1,2}/\d{4}|(?:18|19|20)\d{2})\b'

df['Born_Date'] = df['Born'].str.extract(date_pattern, flags=re.IGNORECASE)
```

```python
df.loc[~df['Born_Date'].str.contains(r'.*\d.*', regex=True, na=True)]
```

```python
df['Born_Date'] = pd.to_datetime(df['Born_Date'], errors='coerce')
```

```python
df.Born_Date
```

```python
df['Died_Date'] = df['Died'].str.extract(date_pattern, flags=re.IGNORECASE)
```

```python
df['Died_Date'] = pd.to_datetime(df['Died_Date'], format="mixed", errors='coerce')
```

```python
df.head(5)
```

```python
df[['Born_City','Born_Region','Born_Country']] =  df['Born'].str.extract(r'in\s+([^,]+),\s+([^(]+)\s+\(([^)]+)\)',  expand=True)
df.head()
```
